Observations

The dataset has 34 columns, many of which contain either non-null values for all rows or are completely null. Here’s an overview of how to clean this dataset:

- Columns with non-null values: These include IndicatorCode, Indicator, ValueType, ParentLocationCode, ParentLocation, Location type,     SpatialDimValueCode, Location, Period type, Period, FactValueNumeric, Value, Language, and DateModified.
  
- Completely null columns: These include Dim3 type, Dim3, Dim3ValueCode, DataSourceDimValueCode, DataSource, FactValueNumericPrefix,      FactValueUoM, FactValueNumericLowPrefix, FactValueNumericLow, FactValueNumericHighPrefix, FactValueNumericHigh,                         FactValueTranslationID, and FactComments.

Cleaning Steps
1. Drop columns with 100% missing values: Columns like Dim3 type, Dim3, and others that have no data should be dropped.

2. Standardize column names

4. Verify relevant columns

```python
# Identify columns with 100% missing values
empty_columns = df4.columns[df4.isnull().all()]
print("Columns with 100% missing values:", empty_columns)
```

```python
# Drop unnecessary columns
columns_to_drop = [
    'Dim3', 'Dim3 type', 'Dim3ValueCode', 'DataSourceDimValueCode', 
    'DataSource', 'FactValueNumericPrefix', 'FactValueUoM', 
    'FactValueNumericLowPrefix', 'FactValueNumericLow', 
    'FactValueNumericHighPrefix', 'FactValueNumericHigh', 
    'FactValueTranslationID', 'FactComments', 'DateModified', 
    'IsLatestYear', 'Dim1', 'Dim1ValueCode', 'Dim2', 'Dim2ValueCode', 
    'Language'
]
df4_cleaned = df4.drop(columns=columns_to_drop)
```

```python
# Standardize column names
df4_cleaned.columns = df4_cleaned.columns.str.strip().str.lower().str.replace(' ', '_')

# Ensure numeric columns are of the correct type
numeric_columns = ['factvaluenumeric', 'value']
df4_cleaned[numeric_columns] = df4_cleaned[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
print(df4_cleaned.isnull().sum())
```

```python
print(df4_cleaned.info())
```

```python
# Save the cleaned dataset to a CSV file
df4_cleaned.to_csv('df4_cleaned.csv', index=False)
```

```python
#Steps to Clean Dataset 7:
```

```python
#Print all the column names in df7 to confirm the exact name.
print(df7.columns.tolist())
```

```python
# Standardize column names: convert to lowercase and replace spaces and special characters with underscores
df7.columns = df7.columns.str.lower()  # Convert to lowercase
df7.columns = df7.columns.str.replace(' ', '_')  # Replace spaces with underscores
df7.columns = df7.columns.str.replace('%', 'pct')  # Replace '%' with 'pct'
df7.columns = df7.columns.str.replace('(', '')  # Remove '('
df7.columns = df7.columns.str.replace(')', '')  # Remove ')'

# Print standardized column names
print(df7.columns)
```

```python
# Impute missing values with the median for relevant columns in df7
df7['bcg_pct_of_one-year-olds_immunized'].fillna(df7['bcg_pct_of_one-year-olds_immunized'].median(), inplace=True)
df7['hepb3_pct_of_one-year-olds_immunized'].fillna(df7['hepb3_pct_of_one-year-olds_immunized'].median(), inplace=True)
df7['hib3_pct_of_one-year-olds_immunized'].fillna(df7['hib3_pct_of_one-year-olds_immunized'].median(), inplace=True)
df7['ipv1_pct_of_one-year-olds_immunized'].fillna(df7['ipv1_pct_of_one-year-olds_immunized'].median(), inplace=True)
df7['mcv1_pct_of_one-year-olds_immunized'].fillna(df7['mcv1_pct_of_one-year-olds_immunized'].median(), inplace=True)
df7['pcv3_pct_of_one-year-olds_immunized'].fillna(df7['pcv3_pct_of_one-year-olds_immunized'].median(), inplace=True)
df7['pol3_pct_of_one-year-olds_immunized'].fillna(df7['pol3_pct_of_one-year-olds_immunized'].median(), inplace=True)
df7['rcv1_pct_of_one-year-olds_immunized'].fillna(df7['rcv1_pct_of_one-year-olds_immunized'].median(), inplace=True)
df7['rotac_pct_of_one-year-olds_immunized'].fillna(df7['rotac_pct_of_one-year-olds_immunized'].median(), inplace=True)
df7['yfv_pct_of_one-year-olds_immunized'].fillna(df7['yfv_pct_of_one-year-olds_immunized'].median(), inplace=True)
df7['dtp3_pct_of_one-year-olds_immunized'].fillna(df7['dtp3_pct_of_one-year-olds_immunized'].median(), inplace=True)

# Drop the 'code' column
df7.drop(columns=['code'], inplace=True)

# Check for remaining missing values
print(df7.isnull().sum())
```

```python
# Save the cleaned dataframe to a CSV file
df7.to_csv('df7_cleaned.csv', index=False)
```
